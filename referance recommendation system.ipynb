```python
def SepGen(text):
    L=[]#deffining an empty list L
    counter=0#to take only top 3 cast
    for i in ast.literal_eval(text):
        if counter<3:
            L.append(i['name'])
            counter+=1
        return L
```

```python
movies['cast'] = movies['cast'].apply(SepGen)
movies.head()
```

```python
movies['crew'].head()
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
def collapse(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```
